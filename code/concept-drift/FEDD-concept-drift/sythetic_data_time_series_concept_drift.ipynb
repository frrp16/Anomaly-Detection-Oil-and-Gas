```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Linear TS

```python
linear = pd.read_csv('../../Documents/phd_related/ing/data_monitoring/synthetic_datasets_time_series/lin-1.csv', header = None)
linear
```

```python
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['figure.figsize'] = (25,6)

plt.plot(linear)
plt.title('Linear Time Series')
plt.xlabel('Timestamp')
plt.ylabel('Value')
for i in range(0, 20001, 2000):
    if(i == 0 or i == 20000):
        plt.axvline(x = i, color = 'b')
    else:
        plt.axvline(x = i, color = 'r')
plt.axvspan(12000, 14000, facecolor = 'y', alpha = 0.2)
```

# Hybrid TS

```python
hybrid = pd.read_csv('../../Documents/phd_related/ing/data_monitoring/synthetic_datasets_time_series/hib-1.csv', header = None)
hybrid
```

```python
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['figure.figsize'] = (25,6)
plt.rc('legend',fontsize=16) 

plt.plot(hybrid)
plt.title('Hybrid Time Series')
plt.xlabel('Timestamp')
plt.ylabel('Value')
for i in range(0, 20001, 2000):
    if(i == 0 or i == 20000):
        plt.axvline(x = i, color = 'b')
    else:
        plt.axvline(x = i, color = 'r')

label = 'Linear'
for i in range(0, 20001, 6000):
    plt.axvspan(i, i+2000, facecolor = 'y', alpha = 0.2, label = label)
    label = "_nolegend_"

label = 'Seasonal'
for i in range(2000, 18001, 6000):
    plt.axvspan(i, i+2000, facecolor = 'g', alpha = 0.2, label = label)
    label = "_nolegend_"

label = 'Non-Linear'
for i in range(4000, 16001, 6000):
    plt.axvspan(i, i+2000, facecolor = 'pink', alpha = 0.2, label = label)
    label = "_nolegend_"

plt.legend()
```
